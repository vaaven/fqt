```python
from fqt import FastQuantileTransformer
from sklearn.preprocessing import QuantileTransformer
import numpy as np
import timeit
import warnings
warnings.filterwarnings("ignore")
```

```python
shape = (1000000, 100)
data = np.random.random(shape)
data
```

```python
number_of_runs = 5
```

```python
timeit.timeit(lambda: QuantileTransformer().fit_transform(data), number=number_of_runs) / number_of_runs
```

```python
timeit.timeit(lambda: QuantileTransformer(subsample=100000).fit_transform(data), number=number_of_runs) / number_of_runs
```

```python
timeit.timeit(lambda: FastQuantileTransformer().fit_transform(data), number=number_of_runs) / number_of_runs
```

```python
timeit.timeit(lambda: FastQuantileTransformer(subsample=100000).fit_transform(data), number=number_of_runs) / number_of_runs
```

```python
timeit.timeit(lambda: FastQuantileTransformer(noise_distribution='uniform').fit_transform(data), number=number_of_runs) / number_of_runs
```
